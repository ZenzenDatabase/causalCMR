```python
import json

def read_lists_from_file(filename):
    with open(filename, 'r') as file:
        loaded_data = json.load(file)
    return loaded_data

loaded_loss_lists = read_lists_from_file("loss_values.json")
loaded_map_lists  = read_lists_from_file("map_values.json")


```

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

epochs = np.arange(1, 1001)

plt.figure(figsize=(8, 6))
plt.plot(epochs, loaded_loss_lists["baseline"], marker='s', label='Baseline', color='pink', markersize=0.5)
plt.plot(epochs, loaded_loss_lists["triplet_300_loss_values"], marker='s', label='Epochs=300', color='orange', markersize=0.5)
plt.plot(epochs, loaded_loss_lists["triplet_400_loss_values"], marker='^', label='Epochs=400', color='red', markersize=0.5)
plt.plot(epochs, loaded_loss_lists["triplet_500_loss_values"], marker='s', label='Epochs=500', color='blue', markersize=0.5)
plt.plot(epochs, loaded_loss_lists["triplet_600_loss_values"], marker='s', label='Epochs=600', color='purple', markersize=0.5)
plt.plot(epochs, loaded_loss_lists["triplet_700_loss_values"], marker='^', label='Epochs=700', color='green', markersize=0.5)
plt.plot(epochs, loaded_loss_lists["triplet_800_loss_values"], marker='s', label='Epochs=800', color='black', markersize=0.5)
plt.plot(epochs, loaded_loss_lists["triplet_900_loss_values"], marker='^', label='Epochs=900', color='yellow', markersize=0.5)

plt.xlabel('epochs', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.legend(loc='upper left', fontsize=12, bbox_to_anchor=(1, 0.7))
plt.grid(True)
plt.show()
```
